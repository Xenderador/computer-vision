```python
import cv2
import matplotlib.pyplot as plt

# Read an image from disk
img = cv2.imread("tomato.jpeg", cv2.IMREAD_COLOR)

# Convert BGR color to RGB color format
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image using plt.imshow()
plt.imshow(RGB_img)

# Read the image in grayscale mode
Gray_img = cv2.imread('tomato.jpeg', cv2.IMREAD_GRAYSCALE)

# Filename to save the grayscale image
filename = 'Gray_image.jpg'

# Save the grayscale image
cv2.imwrite(filename, Gray_img)

# Create a GUI window to display the original image
cv2.imshow("Original_image", img)

# Display the grayscale image
cv2.imshow('Gray_Scale_img', Gray_img)

# To hold the window on the screen, use cv2.waitKey method
# Once it detects the close input, it will release the control
# to the next line
# The first parameter is for holding the screen for specified milliseconds
# If you pass 0 as a parameter, it will hold the screen until the user closes it.
cv2.waitKey(0)

# Remove/delete the created GUI window from the screen and memory
cv2.destroyAllWindows()

# Create a VideoCapture object and read from the input video file
cap = cv2.VideoCapture('Sample.mp4')

# Check if the camera opened successfully
if not cap.isOpened():
    print("Error opening video file")

# Read and display each frame until the video is completed
while cap.isOpened():
    # Capture frame-by-frame
    ret, frame = cap.read()
    if ret:
        # Display the resulting frame
        cv2.imshow('Frame', frame)

        # Press 'Q' on the keyboard to exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

# Release the video capture object and close all windows 
cap.release()
cv2.destroyAllWindows()
```
